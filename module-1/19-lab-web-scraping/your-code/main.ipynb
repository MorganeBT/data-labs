````python
# Find out the html tag and class names used for the developer names. You can achieve this using Chrome DevTools.
# Use BeautifulSoup to extract all the html elements that contain the developer names.
h1tags = soup.find_all('h1',{'class':'h3 lh-condensed'})
````

````python
#Use string manipulation techniques to replace whitespaces and linebreaks (i.e. \n) in the text of each html element.

names = [] 

for h1tag in h1tags:
    a_tag = h1tag.find('a')
    #print(a_tag)
    user_name = a_tag.get('href')[1:] # la fonction .get permet de trouver un attribut dans un tag (librairie Beautiful Soup)
    full_name = a_tag.text
    names.append((user_name, full_name))
print(names)
````

````python
#Use a list to store the clean names.
final_list = [user[0] + " (" + user[1] + ")" for user in names]
````

````python
#Print the list of names.
print(final_list)
````
