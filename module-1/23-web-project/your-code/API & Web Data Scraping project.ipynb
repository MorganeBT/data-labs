```python
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
from bs4 import BeautifulSoup
```

```python
""" => Obtain data from an API using Python.
For the API portion of the project will need to make calls to your chosen API, 
successfully obtain a response, request data, convert it into a Pandas data frame, 
and export it as a CSV file
"""

url = 'https://itunes.apple.com/search?term=jack+johnson'
```

```python
# request data
r = requests.get(url)
```

```python
results = r.json()
results
```

```python
# convert it into a Pandas data frame, 2 columns in 34 columns

df = json_normalize(results, record_path=['results'], meta=['resultCount'])
df
```

```python
# export it as a CSV file
df.to_csv('jackjonhson.txt', index = False)
```

```python
""" => Scrape and clean HTML from a web page using Python.
For the web scraping portion of the project, 
you will need to scrape the HTML from your chosen page, 
parse the HTML to extract the necessary information, 
and either save the results to a text (txt) file if it is text 
or into a CSV file if it is tabular data.
"""

url2 = 'https://fr.wikipedia.org/wiki/Féminisme'
```

```python
# scrape the HTML from your chosen page
html = requests.get(url2).content
```

```python
soup = BeautifulSoup(html, "lxml")
soup
```

```python
# parse the HTML to extract the necessary information : links of pictures

image_tag = ['img']
image_tags = [image for image in soup.find_all(image_tag)]
image_tags
```

```python
imagelist = soup.find_all('a', {'class': 'image'})
imagelist
```

```python
images = [a.get('href')for a in imagelist]
images
```

```python
# either save the results to a text (txt) file if it is text
images = pd.DataFrame(images)
```

```python
images.to_csv('images_links.txt', sep=',', index = False)
```
