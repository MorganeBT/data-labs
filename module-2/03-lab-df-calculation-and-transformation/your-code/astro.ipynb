```python
import pandas as pd
import numpy as np
import os
from scipy import stats
from scipy.stats import chi2_contingency
```

```python
astro_metier=pd.read_csv('astro_metier.csv')
astro_metier
```

```python
contingency_table = pd.crosstab(astro_metier.element, astro_metier.categorie)
contingency_table
```

```python
#scipy.stats.chi2 = <scipy.stats._continuous_distns.chi2_gen object>
chi2, p, dof, ex = chi2_contingency(contingency_table, correction=False)
```
