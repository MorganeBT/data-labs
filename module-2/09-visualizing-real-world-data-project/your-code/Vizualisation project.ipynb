```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('World Happiness 2017.csv')
data.tail(10)
#see more: https://worldhappiness.report/ed/2017/
```

```python
# What are the Whiskers values?
# It's a referral to the values you'd get if you plotted the values in a box plot which has whiskers on each end of the plot.
# So I choose to drop them
data1 = data.drop(columns =["Whisker.high", "Whisker.low","Happiness.Rank","Dystopia.Residual"])
data1.rename(columns={'Family':'Social support'}, inplace=True)
data1.head()
```

```python
data1.describe()
```

```python
"""
one scatter plot, one histogram, one box plot and one bar graph 
(you can add more than one visualization of each type of you choose). 
Graphs should contain the proper labeling of the x and y axis when appropriate as well as a title for the graph.
Explain what insight or information is inferred from these visualizations.
"""
```

```python
# I am looking at eventual correlation between each serie.
sns.heatmap(data1.corr(), cmap='winter')
sns.set(rc={'figure.figsize':(10,5)})
plt.title('Correlation matrix')
```

```python
data1[["Social support", "Generosity"]].agg('mean').plot.bar(figsize = (10,5), color = "g").set_title('Do not mix Social support VS Generosity')
```

```python
cols=['Freedom','Health..Life.Expectancy.','Generosity','Happiness.Score']
df=data1[cols]
fig,axes = plt.subplots(figsize = (16,8), nrows = 1, ncols = 3)
for i in range(3):
    plt.sca(axes[i])
    plt.scatter(df.iloc[:,i],df['Happiness.Score'], alpha = 0.7, color = 'r')
    plt.xlabel(df.iloc[:,i].name)
    plt.ylabel('Happiness.Score')
plt.title('Generosity is not considered as a indication of happiness, \n whereas freedom, health & life expectancy are')
```

```python
sns.scatterplot(data1['Social support'], data1['Happiness.Score']).set_title("Where social support is appreciated, happiness is high")
```

```python
Generosity = data1.sort_values(['Generosity'], ascending=False).set_index('Country')
#Generosity
```

```python
Generosity['Happiness.Score'].head(5).plot.barh(title='The 5 countries most rated in Generosity are not the happier', figsize =(10,5), color= "orange")
```

```python
Generosity = data1.sort_values(['Generosity'], ascending=False).drop(columns=["Happiness.Score"])
Generosity.boxplot(figsize=(20,10))
```

```python
Corruption = data1.sort_values(by = 'Trust..Government.Corruption.', ascending=False).drop(columns=["Happiness.Score"])
Corruption.boxplot(figsize=(20,10))
```

```python
sns.distplot(data1['Happiness.Score'], bins=8, color='magenta').set_title('Globally, the world is not that happy')
sns.set(rc={'figure.figsize':(8,8)})
```
