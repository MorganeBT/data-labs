```python
import pandas as pd
```

```python
df = pd.read_csv('data/austin_weather.csv')
```

```python
df
```

```python
df.Events.value_counts()
```

```python
list_of_created_columns = ['Rain', 'Thunderstorm', 'Fog', 'Snow']
for i in list_of_created_columns :
    df[i] = 0
```

```python
for column in list_of_created_columns:
    df[f'{column}'][df.Events.str.contains(f'{column}') == True] = 1 
```
